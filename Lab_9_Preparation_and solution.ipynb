## Question 2

Train two SVM model from `sklearn.svm.SVC` with kernel 'rbf', gamma = 'auto', and C = 10. 
Target column is `y`. 
- Model 1 will be trained on original data (after scaling)
- Model 2 will be trained on the dataset obtained after PCA with 130 components

Split data to Train and Test, 80/20 ratio. 

Report following metrics on the test dataset:
- Confusion matrix
- F1 weighted

Set random_state = 142 on all models and functions with this argument.

How the model tow model results are different?  
Argue in favor and against using PCA to reduce number of features in this scenario. Provide at least two arguments for each. 

### Question 2 Solution

````python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score


scaler = MinMaxScaler()


X = data.drop(columns=['y'])
y = data['y']

""" Split to Train/Test"""
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=142)

""" Scale dataset, exclude y """
scaler.fit(X_train)
X_tr_sc = scaler.transform(X_train)
X_tst_sc = scaler.transform(X_test)

pca = PCA(n_components = 130)
X_tr_reduced = pca.fit_transform(X_tr_sc)
X_tst_reduced = pca.transform(X_tst_sc)

# Train model 1
svc1 = SVC(gamma='auto', kernel='rbf', C=10, random_state=142)
svc1.fit(X_tr_sc, y_train)

"""Score on Testing data """
y_tst_pred = svc1.predict(X_tst_sc)
print("Results of run on full dataset")
print(confusion_matrix(y_test, y_tst_pred))
print("F1 for Test:", f1_score(y_test, y_tst_pred, average='weighted'))


# Train model 2
svc2 = SVC(gamma='auto', kernel='rbf', C=10, random_state=142)
svc2.fit(X_tr_reduced, y_train)

"""Score on Testing data """
y_tst_pred = svc2.predict(X_tst_reduced)
print("\nResults of run on PCA reduced dataset")
print(confusion_matrix(y_test, y_tst_pred))
print("F1 for Test:", f1_score(y_test, y_tst_pred, average='weighted'))
````

The results of two models are close.

- In favor of PCA
    - We get very similar results with number of features reduced from 200 to 125
    - Model training and potentially model size are improved
    
- Against PCA:
    - It is much harder to explain the model now
    - We need one more step in production pipeline to run PCA before we can score

### Q3 solution

````python
import h2o
from h2o.estimators import H2OGeneralizedLowRankEstimator

h2o.init()
h2o.no_progress()
````

````python
# Import Data
h_data = h2o.H2OFrame(data)
h_data.head()
````

````python
glrm_cols = [x for x in h_data.columns if x not in ["y"]]


rank2_model = H2OGeneralizedLowRankEstimator(k=2,
                                             model_id = "rank2", 
                                             seed = 1234, 
                                             loss="Absolute",
                                             transform="standardize",
                                             representation_name="rank2_full")
rank2_model.train(training_frame=h_data, x=glrm_cols);
````

````python
X = h2o.get_frame("rank2_full")
Y = rank2_model.archetypes()
````

````python
print(X.shape)
print(len(Y), len(Y[0]))
````

````python
import pandas as pd

archetypes_cols = rank2_model._model_json["output"]["archetypes"].as_data_frame().reset_index(drop=True).columns.to_list()
archetypes_y = pd.DataFrame({'Arch1':Y[0],'Arch2':Y[1], 'attribute':archetypes_cols[1:]})
archetypes_y.head(n=20)
````

````python
from matplotlib.pyplot import figure
import matplotlib.pyplot as pl
import random

pl.rcParams['figure.figsize'] = [15, 10]

ax1 = archetypes_y.plot.scatter(x='Arch1',
                                y='Arch2',
                                c='DarkBlue')
for i, txt in enumerate(archetypes_cols[1:]):
    if txt in ["col_191","col_145", "col_84","col_99","col_60","col_49"]:
        ax1.annotate(txt, (Y[0][i]+random.uniform(-0.1,0.1), Y[1][i]+random.uniform(-0.1,0.1)))


pl.show()
````
